````python
def request_sreality(category_main, category_type, locality_region, page):
    mains = {'flat':1, 'house':2, 'land':3 }
    types = {'sell':1,'rent':2}
    locations = {'Praha':10,'Brno':14}
    
    url_template = 'https://www.sreality.cz/api/cs/v2/estates?category_main_cb={cm}&category_type_cb={ct}&locality_region_id={lr}&per_page60&page={page}'
    
    
    try:
        url = url_template.format(
            cm=mains[category_main],
            ct=types[category_type],
            lr=locations[locality_region],
            page=page
        )
        r = requests.get(base_url)
        if r.ok: 
            return r.json()
    except:
        print('There was an error in request')
    

d = request_sreality('land','rent','Brno', 1)
````

### 1c. Combining multiple requests into single df

* Function should parametrize:
    * `start_page` and `end_page`
    * request parameters
* construct a list of individual request dfs
* then feed it into `pd.concat` function

````python
list(range(1,10))
````

````python
def multiple_sreality_requests(start_page,end_page,category_main='flat', category_type='sell', locality_region='Praha'):
    return pd.concat([
        request_sreality(category_main=category_main, category_type=category_type, locality_region=locality_region,page=i) 
        for i in range(start_page,end_page+1)
    ])

raw = multiple_sreality_requests(0,4)
raw.head()
````
